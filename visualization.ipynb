## Filter out HDBs that are not sold

```python
df_processed = df[(df['1room_sold'] > 0 ) | (df['2room_sold'] > 0 ) | (df['3room_sold'] > 0 ) | (df['4room_sold'] > 0 ) | (df['5room_sold'] > 0 ) | (df['exec_sold'] > 0 ) | (df['multigen_sold'] > 0 ) | (df['studio_apartment_sold'] > 0 )]

df_processed
```

```python
# create address list to be inputted into API
df = df_processed
df['Address'] = df['blk_no'] + " " + df['street']
addresslist = list(df['Address'])
addresslist
```

```python
def getAllData(address):
    req = requests.get('https://developers.onemap.sg/commonapi/search?searchVal='+address+'&returnGeom=Y&getAddrDetails=Y&pageNum=1')
    resultsdict = eval(req.text)
    if len(resultsdict['results'])>0:
        return resultsdict
    else:
        pass
```

```python
getAllData(addresslist[0])
```

```python
def getData(address):
    req = requests.get('https://developers.onemap.sg/commonapi/search?searchVal='+address+'&returnGeom=Y&getAddrDetails=Y&pageNum=1')
    resultsdict = eval(req.text)
    if len(resultsdict['results'])>0:
        return resultsdict['results'][0]['BLK_NO'], resultsdict['results'][0]['ROAD_NAME'], resultsdict['results'][0]['BUILDING'], resultsdict['results'][0]['ADDRESS'], resultsdict['results'][0]['POSTAL'], resultsdict['results'][0]['LATITUDE'], resultsdict['results'][0]['LONGITUDE']
    else:
        pass
```

```python
getData(addresslist[0])
```

```python
import time
from tqdm import tqdm
coordinateslist= []
count = 0
failed_count = 0
for address in tqdm(addresslist):
    try:
        if len(getData(address))>0:
            count = count + 1
            #print('Extracting',count,'out of',len(addresslist),'addresses')
            coordinateslist.append(getData(address))
    except:
        count = count + 1           
        failed_count = failed_count + 1
        #print('Failed to extract',count,'out of',len(addresslist),'addresses')
        coordinateslist.append(None)

print('Total Number of Addresses parsed',count)
print('Total Number of Addresses With No Coordinates',failed_count)
```

```python
df_coordinates = pd.DataFrame(coordinateslist)
df_combined = df.join(df_coordinates)
df_combined  = df_combined .rename(columns={0:'BLK_NO', 
                                            1:'ROAD_NAME',
                                            2:'BUILDING',
                                            3:'ADDRESS',
                                            4:'POSTAL',
                                            5:'LATITUDE',
                                            6:'LONGITUDE',
                                            })
df_combined
```

```python
df_combined.to_csv("hdb_property_processed.csv",index=False)
```

```python
from geopandas import GeoDataFrame
from shapely import geometry


geometry = [geometry.Point(xy) for xy in zip(df.LONGITUDE, df.LATITUDE)]
gdf = GeoDataFrame(df, crs="EPSG:3857", geometry=geometry)
gdf
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
# import our image 
singapore_img = mpimg.imread('singapore.png')
# plot the data
ax = df.plot(
    kind="scatter", 
    x="LONGITUDE", 
    y="LATITUDE", 
    figsize=(20,14),
    c="year_completed", 
    cmap=plt.get_cmap("jet"),
    colorbar=True, 
    alpha=0.4,
)
# use our map with it's bounding coordinates
plt.imshow(singapore_img, extent=[103.5,104,1.15, 1.50], alpha=0.5)            
# add axis labels
plt.ylabel("Latitude", fontsize=20)
plt.xlabel("Longitude", fontsize=20)
# set the min/max axis values - these must be the same as above
plt.ylim(1.15, 1.50)
plt.xlim(103.5, 104)
plt.legend(fontsize=20)
plt.show()
```
